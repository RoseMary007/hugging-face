```python
import torch
import torch.nn.functional as F
```

```python
sentence="I learn machine learning"
token_ids=tokenizer(sentence,return_tensors="pt").input_ids
outputs=gpt2(token_ids).logits[0,-1]
final_logits=torch.topk(outputs,5)


for index in final_logits.indices:
  print(tokenizer.decode(index))
```

```python
def greedy_decode(logits):

  return torch.argmax(logits,dim=-1)
tokenizer.decode(greedy_decode(outputs))
```

```python
def tok_k_sampling(logits,k=50):
  """
  keeps inly top k logits.normalize them into probablilty.
  them sample one token from the filtered distribution
  """
  values,indices=torch.topk(logits,k)
  probs=F.softmax(values,dim=-1)
  sampled=torch.multinomial(probs,1)
  return indices[sampled]

def top_p_sampling(logits,p=0.9):
  """sort tokens by prob. keep smallest number whose cumulative prob
  exceeds treshold p, then sample one token"""

  sorted_logits,sorted_indices=torch.sort(logits,descending=True)
  sorted_probs=F.softmax(sorted_logits,dim=-1)
  cumulative_probs=sorted_probs.cumsum(dim=-1)


  mask = cumulative_probs > p
  sorted_logits[mask] = float("-inf")
  filtered_probs=F.softmax(sorted_logits,dim=-1)
  sampled=torch.multinomial(filtered_probs,1)
  return sorted_indices[sampled]
```

```python
tokenizer.decode(tok_k_sampling(outputs))
```

```python
outputs
```

```python
tokenizer.decode(top_p_sampling(outputs,p=0.9))
```

```python
#temp sampling

def temperature_sampling(logits,temperature=1.0):
  """
  scale logits by temperature before sampling
  lower temperature values - sharper probability distribution
  higher temperture values -flatened " "

  """

  scaled=logits/temperature
  probs=F.softmax(scaled,dim=-1)
  return torch.multinomial(probs,1)
```

```python
temperature_sampling(outputs,temperature=0.9)
```

```python
tokenizer.decode(temperature_sampling(outputs,temperature=1.3))
```

```python
def random_sampling(logits):
   """
   sample directly from softmax distribution without any filtering
   """
   probs=F.softmax(logits,dim=-1)
   return torch.multinomial(probs,1)
```

```python
random_sampling(outputs)
```

```python
tokenizer.decode(random_sampling(outputs))
```

```python
sentence="I am really happy because every"
inputs=tokenizer(sentence,return_tensors="pt")
outputs=gpt2(**inputs)
logits=outputs.logits[0,-1]

print(f"greedy decode ",tokenizer.decode([greedy_decode(logits)]))
print(f"top k samplimng ",tokenizer.decode(tok_k_sampling(logits,k=10)))
print(f"top p sampling ",tokenizer.decode(top_p_sampling(logits,p=0.9)))
print(f"temp ",tokenizer.decode(temperature_sampling(logits,temperature=0.7)))
print(f"randon ",tokenizer.decode(random_sampling(logits)))
```

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)


# → "Bonjour, comment allez-vous ?"
```

```python
text = ["Hello, how are you?"]
tokens = tokenizer(text, return_tensors="pt", padding=True)
translated = model.generate(**tokens)

print(tokenizer.decode(translated[0], skip_special_tokens=True))
```
