```python
from transformers import AutoImageProcessor,AutoModelForImageClassification
processor=AutoImageProcessor.from_pretrained("microsoft/resnet-18")
model=AutoModelForImageClassification.from_pretrained("microsoft/resnet-18")
```

```python
model
```

```python
def classify_image(image):
  image=processor(image,return_tensors="pt")["pixel_values"]

  logits=model(image).logits
  predicted_label =logits.argmax(-1).item()
  return model.config.id2label[predicted_label]
```

```python
interface=gr.Interface(fn=classify_image,inputs=gr.Image(),outputs=gr.Label(),description="uploaded an image")
interface.launch(share=True)
```

```python
from transformers import pipeline
```

```python
def validate_even(num):
  return num %2==0
```

```python
validate_even(4)
```

```python
def add_even(a,b):
  if not validate_even(a):
    raise gr.Error("first value is not even")
  if not validate_even(b):
    raise gr.Error("second value is not even")
  else:
    return a+b
```

```python
iface=gr.Interface(fn=add_even,inputs=[gr.Number(),gr.Number()],outputs=gr.Number())
iface.launch(share=True)
```

```python
def add_even(a,b):
  if not validate_even(a):
    raise gr.Warning("first value is not even")
  if not validate_even(b):
    raise gr.Warning("second value is not even")
  return a+b
```

```python
iface=gr.Interface(fn=add_even,inputs=[gr.Number(),gr.Number()],outputs=gr.Number())
iface.launch(share=True)
```

```python
def add_even(a,b):
  if not validate_even(a):
      gr.Info("first value is not even")
  if not validate_even(b):
     gr.Info("second value is not even")
  return a+b
```

```python
iface=gr.Interface(fn=add_even,inputs=[gr.Number(),gr.Number()],outputs=gr.Number())
iface.launch(share=True)
```
